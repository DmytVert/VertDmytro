# II. Цілі числа та числа з рухомою комою

4. Створити дві змінні, що посилаються на будь-які два цілих числа та продемонструвати такі арифметичні операції: додавання, віднімання, ділення, множення, ділення без залишку, ділення по модулю, приведення до ступеню. Всі результати операцій вивести на екран.

```python
# Створення двох змінних, що посилаються на цілі числа
x = 15
y = 4

# Додавання
addition_result = x + y
print(f"Додавання: {addition_result}")

# Віднімання
subtraction_result = x - y
print(f"Віднімання: {subtraction_result}")

# Множення
multiplication_result = x * y
print(f"Множення: {multiplication_result}")

# Ділення
division_result = x / y
print(f"Ділення: {division_result}")

# Ділення без залишку (цілочисельне ділення)
integer_division_result = x // y
print(f"Ділення без залишку: {integer_division_result}")

# Ділення по модулю
modulus_result = x % y
print(f"Ділення по модулю: {modulus_result}")

# Піднесення до ступеня
exponentiation_result = x ** y
print(f"Піднесення до ступеня: {exponentiation_result}")
```

```python
# Створення двох змінних, що посилаються на числа з рухомою комою
a = 7.5
b = 2.0

# Додавання
addition_result = a + b
print(f"Додавання: {addition_result}")

# Віднімання
subtraction_result = a - b
print(f"Віднімання: {subtraction_result}")

# Множення
multiplication_result = a * b
print(f"Множення: {multiplication_result}")

# Ділення
division_result = a / b
print(f"Ділення: {division_result}")

# Ділення без залишку (цілочисельне ділення) - не застосовується до чисел з рухомою комою
# Ця операція зруйнує числа з рухомою комою, тому її краще уникати для таких чисел

# Ділення по модулю - також не застосовується до чисел з рухомою комою
# Так само, як і цілочисельне ділення, це операція непризначена для чисел з рухомою комою

# Піднесення до ступеня
exponentiation_result = a ** b
print(f"Піднесення до ступеня: {exponentiation_result}")
```

6. Спробуйте проробити всі ті самі арифметичні операції над двома змінними, що посилаються на різні булеві значення. Прокоментуйте, чому, на Вашу думку, Ви отримали саме такі результати.

```python
# Створення двох змінних, що посилаються на булеві значення
bool_var1 = True
bool_var2 = False

# Додавання
addition_result = bool_var1 + bool_var2
print(f"Додавання: {addition_result}")

# Віднімання
subtraction_result = bool_var1 - bool_var2
print(f"Віднімання: {subtraction_result}")

# Множення
multiplication_result = bool_var1 * bool_var2
print(f"Множення: {multiplication_result}")

# Ділення
division_result = bool_var1 / bool_var2
print(f"Ділення: {division_result}")
```

*Додавання: 1 + 0 = 1
Віднімання: 1 - 0 = 1
Множення: 1 * 0 = 0
Ділення: 1 / 0 - це спричинить помилку ділення на нуль, оскільки False (0) використовується для представлення False у булевому контексті, і ділення на 0 не є допустимим.
Важливо пам'ятати, що це автоматичне конвертування булевих значень у цілі числа відбувається в контексті арифметичних операцій, і в інших випадках, наприклад, порівняння (<, >, ==), булеві значення розглядаються як булеві і не конвертуються в цілі числа.*

9. Створіть два списка та за допомогою спеціального методу додайте другий з них в якості останнього елемента першого.

```python
#Використання операції +=
list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Використання операції += для додавання елементів з другого списку в кінець першого
list1 += list2

# Виведення результату
print(f"Об'єднаний список (за допомогою +=): {list1}")
```

10. Створіть два списка та за допомогою спеціального методу "розширте" перший (додайте всі елементи другого в кінець першого списку). Напишіть, чим відрізняються методи в завданні 9 та 10.

```python
#Використання методу extend()
list1 = [1, 2, 3]
list2 = [4, 5, 6]

# Використання методу extend() для додавання елементів з другого списку в кінець першого
list1.extend(list2)

# Виведення результату
print(f"Об'єднаний список (за допомогою extend()): {list1}")
```

*Основна відмінність полягає в тому, що операція += є більш компактним способом виконання тієї ж самої дії. Обидва підходи, += та extend(), призводять до однакового результату - об'єднаного списку. Однак використання += може бути сприйняте як більш зручний та краще читається спосіб, особливо для коротких операцій.*

17. Створіть список з 6ти елементів. Отримайте третій елемент за допомогою двох різних індексів.

18. Створіть список, де елементами цього списку також є списки. Отримай перший елемент з останнього рядка та виведи значення на екран.

```python
# Створення списку, де елементами є списки
nested_list = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Отримання першого елемента з останнього рядка
first_element_last_row = nested_list[-1][0]

# Виведення результату
print("Вкладений список:")
for row in nested_list:
    print(row)

print("\nПерший елемент з останнього рядка:", first_element_last_row)
```
